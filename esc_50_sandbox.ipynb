```python
import pandas as pd

meta = pd.read_csv('data/ESC-50-master/meta/esc50.csv')

animal_classes = [
    'dog',
    'chirping_birds',
    'crow',
    'sheep',
    'frog',
    'cow',
    'hen',
    'pig',
    'rooster',
    'cat',
    'crickets',
]

animal_subset = meta[meta['category'].isin(animal_classes)]
animal_subset = animal_subset.reset_index(drop=True)

target_dict = {}
for animal in animal_classes:
    target_dict[animal] = int(animal_subset[animal_subset['category'] == animal]['target'].iloc[0])

animal_subset
```
